```python
#fig, ax = plt.subplots(1)
fig, ax = plt.subplots(2,3)
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)


ax[0,0].scatter(x,y, s=1)
ax[0,0].plot(x,x)
ax[0,1].hist(y)

ax[0,0].set_title('Plot1')
ax[0,0].set_xlabel('My X Axis')
plt.xlabel('Global X')
plt.suptitle('Titolo!')
plt.tight_layout()

plt.savefig('myplot.pdf',)
plt.savefig('myplot.png', dpi = 600)
```
